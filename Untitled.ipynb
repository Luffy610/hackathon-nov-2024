```python
from llm_components.manager import Manager, llm
```

```python
import pandas as pd
```

```python
df = pd.read_excel("data/customerSegment.xlsx")
```

```python
df.dropna(how="all", axis=1, inplace=True)
```

```python
text_gen = llm(
    provider="openai",
    api_type="azure",
    azure_endpoint="https://genai-openai-aialchemists1.openai.azure.com/openai/deployments/gpt-35-turbo/chat/completions?api-version=2024-08-01-preview",
    api_key="0bed3fff0b954af1b2a750704d335499",
    api_version="2023-07-01-preview"
)
```

```python
bot = Manager(text_gen=text_gen)
```

```python
summary = bot.summarize(data=df, summary_method='llm')
```

```python
summary
```

```python
goals = bot.goals(summary=summary, persona=None, no_of_goals=10)
```

```python
goals
```

```python
chart_code = bot.chart_generator(goal=goals[0], summary=summary)
```

```python
chart_code
```

```python
chart_response = bot.execute(code_specs=chart_code, data=df, summary=summary, library="matplotlib", return_error=False)
```

```python
base64_data = chart_response[0].raster
```
